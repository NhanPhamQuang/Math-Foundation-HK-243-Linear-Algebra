```python
import torch
```

```python
# 1. Geometry of Vectors

v = torch.tensor([2., 3.])
w = torch.tensor([1., -1.])
add = v + w
sub = v - w
print("v + w =", add)
print("v - w =", sub)
```

```python
# 2. Dot Products and Angles

u = torch.tensor([1., 6.])
v = torch.tensor([2., 4.])
dot = torch.dot(u, v)
norm_u = torch.norm(u)
norm_v = torch.norm(v)
angle_rad = torch.acos(dot / (norm_u * norm_v))
angle_deg = angle_rad * 180 / torch.pi
print("Angle (degrees):", angle_deg.item())
```

```python
# Create tensor in PyTorch
v = torch.tensor([2, 1, 4], dtype=torch.float32)
w = torch.tensor([4, 2, 9], dtype=torch.float32)

# Calculate cosine similarity
cos_sim = torch.dot(v, w) / (torch.norm(v) * torch.norm(w))
print("Cosine similarity:", cos_sim.item())  # Output: 1.0
```

```python
# 3. Hyperplanes

import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-1, 2, 100)
y = 1 - 2 * x  # Hyperplane: 2x + y = 1
plt.plot(x, y, label='2x + y = 1')
plt.scatter([1], [1], c='r', label='Point (1,1) > 1')
plt.scatter([0], [0], c='g', label='Point (0,0) < 1')
plt.axhline(0, color='k', linestyle='--')
plt.axvline(0, color='k', linestyle='--')
plt.legend()
plt.grid()
# plt.show()
```

```python
# 4. Geometry of Linear Transformations

import torch
A = torch.tensor([[1., 4.], [-1., 6.]])
v = torch.tensor([5., -1.])
Av = A @ v
print("A @ v =", Av)
```

```python
# 5. Linear Dependence

# Create tensor in PyTorch
V = torch.tensor([[2, -1], [4, -9]], dtype=torch.float32)

# Calculate rank of the matrix
rank = torch.linalg.matrix_rank(V)
print("Rank:", rank.item())
```

```python
# 6. Rank

# Create tensor in PyTorch
B = torch.tensor([[2, 5], [1, 9]], dtype=torch.float32)

# Calculate rank of the matrix B
rank = torch.linalg.matrix_rank(B)
print("Rank of B:", rank.item())  # Output: 1
```

```python
# 7. Invertibility

A = torch.tensor([[1., 6.], [7., 3.]])
A_inv = torch.inverse(A)
print("A inverse:\n", A_inv)
print("A @ A_inv:\n", A @ A_inv)  # Should be identity matrix
```

```python
# 8. Determinant

A = torch.tensor([[1., 3.], [2., 8]])
det = torch.det(A)
print("Determinant:", det.item())
```

```python
# 9. Tensors and Common Linear Algebra Operations

x = torch.ones(2, 2, 2, 2)  # 4D tensor
a = torch.tensor([[1., 5.], [3., 8.]])  # 2D matrix
y = torch.einsum("ijkl, jk -> il", x, a)
print("Result shape:", y.shape)  # Output: torch.Size([2, 2])
print("Result:\n", y)
```
